```python
import os
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database specified by db_file
    - param db_file: database file
    - return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return conn
```

```python
database_path = os.path.join("cataloqueSqlite.db")
catalogue_path = os.path.join("..", "2_Data_preparation","Final_catalogue.csv")

# create a database connection
connection = create_connection(database_path)
```

```python
# create tables : Method 1
df = pd.read_csv(catalogue_path, sep=';', index_col=0)
df.to_sql('messier', connection, if_exists='replace', index=False)
```
